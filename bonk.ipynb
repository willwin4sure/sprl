```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
win_rates = (1 / (768.0 * 2)) * np.array([
    [0, 55, 44, 38, 49, 38],
    [1481, 0, 596, 508, 581, 578],
    [1492, 940, 0, 730, 749, 699],
    [1498, 1028, 806, 0, 834, 854],
    [1487, 955, 787, 702, 0, 783],
    [1498, 958, 837, 682, 753, 0],
])
```

```python
# make the same plot as above but with axis labels
fig, ax = plt.subplots()
# mask the main diagonal
mask = np.eye(*win_rates.shape, dtype=bool)
sns.heatmap(win_rates, annot=True, fmt=".4f", cmap="YlGnBu", mask=mask, ax=ax)
ax.set_xticklabels(["0", "3", "6", "9", "12", "15"])
ax.set_yticklabels(["0", "3", "6", "9", "12", "15"])
plt.show()
```

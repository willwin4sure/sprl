```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
win_rates = (1 / 768.0) * np.array([
    [0.0, 14.0, 29.5, 20.5, 3.5, 5.5],
    [754.0, 0.0, 94.5, 205.5, 115.5, 91.0],
    [738.5, 673.5, 0.0, 222.5, 199.0, 303.5],
    [747.5, 562.5, 545.5, 0.0, 396.5, 247.5],
    [764.5, 652.5, 569.0, 371.5, 0.0, 250.0],
    [762.5, 677.0, 464.5, 520.5, 518.0, 0.0]
])
```

```python
# make the same plot as above but with axis labels
fig, ax = plt.subplots()
# mask the main diagonal
mask = np.eye(*win_rates.shape, dtype=bool)
sns.heatmap(win_rates, annot=True, fmt=".4f", cmap="YlGnBu", mask=mask, ax=ax)
ax.set_xticklabels(["0", "5", "10", "15", "20", "25"])
ax.set_yticklabels(["0", "5", "10", "15", "20", "25"])
plt.show()
```
